```python
area = np.array([700,800,1000,1500,2000])
price = np.array([40,45,48,60,70])

plt.figure()

plt.title("Area vs Price plot")
plt.xlabel("Area")
plt.ylabel("Price")

plt.plot(area,price)


plt.show()
```

```python
# xlim and ylim
```

```python
area = np.array([700,800,1000,1500,2000])
price = np.array([40,45,48,60,70])

plt.figure()

plt.title("Area vs Price plot")
plt.xlabel("Area")
plt.ylabel("Price")

plt.xlim(400,2500)
plt.ylim(10,100)

plt.plot(area,price)


plt.show()
```

```python
# xticks and yticks
```

```python
area = np.array([700,800,1000,1500,2000])
price = np.array([40,45,48,60,70])

plt.figure()

plt.title("Area vs Price plot")
plt.xlabel("Area")
plt.ylabel("Price")


plt.xticks(rotation=45)
plt.yticks(rotation=90)

plt.plot(area,price)


plt.show()
```

```python
# Multiple plots
```

```python
area = np.array([700,800,1000,1500,2000])
price_mumbai = np.array([40,45,48,60,70])
price_thane = np.array([35,41,42,58,65])

plt.figure(figsize=(10,6))

plt.title("Area vs Price plot")
plt.xlabel("Area")
plt.ylabel("Price")


plt.plot(area,price_mumbai,color="yellow")
plt.plot(area,price_thane,color="red")

plt.legend(labels=["Mumbai","Thane"])


plt.show()
```

## Histogram

<b> Frequency plot for continous data </b>

```python
price = np.array([35,41,45,47,50,51,51,59,59,60,60,61,65,70,75,76,77,78,79,80])
```

```python
# bins = 4
# (80 - 35)/4 = 11


# bin1 -> 35 - 46 -> 3
# bin2 -> 46 - 57 -> 4
# bin3 -> 57 - 68 -> 6 
# bin4 -> 68 - 79 -> 8 
```

```python
plt.hist(price,bins=4)
plt.show()
```

```python
# car dataset
cars = pd.read_csv("cars.csv")
```

```python
cars.head()
```

```python
plt.figure()

plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.hist(cars["price"])


plt.show()
```

```python
price_mean = cars["price"].mean()
price_median = cars["price"].median()
```

```python
plt.figure()

plt.title("Price distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.axvline(price_mean,color="green")
plt.axvline(price_median,color="red")

plt.hist(cars["price"])


plt.show()
```

```python
plt.figure()

plt.title("Width distribution")
plt.xlabel("Width")
plt.ylabel("Frequency")


plt.hist(cars["width"])


plt.show()
```

## Barplot

<b> Frequency plot for categorical plot </b>

```python
place = ["Mumbai","Thane","Kalyan","Panvel"]
no_of_flats = [800,750,900,1000]
```

```python
plt.figure()
plt.bar(place,no_of_flats)
plt.show()
```

```python
plt.figure()
plt.barh(place,no_of_flats)

plt.title("Place vs Flat Frequency")
plt.xlabel("No of flats")
plt.ylabel("Place")

plt.show()
```
